# **L'affichage de l'ensemble de données (6 classes)**

```python
import os
import matplotlib.pyplot as plt
import cv2

base_dir = os.path.join(path, "Faulty_solar_panel")
categories = os.listdir(base_dir)
print("Catégories :", categories)

# Compter le nombre d’images dans chaque classe
for cat in categories:
    folder = os.path.join(base_dir, cat)
    print(f"{cat}: {len(os.listdir(folder))} images")

# Afficher quelques exemples
plt.figure(figsize=(12, 8))
for i, cat in enumerate(categories[:6]):
    folder = os.path.join(base_dir, cat)
    img_name = os.listdir(folder)[0]
    img = cv2.imread(os.path.join(folder, img_name))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(2, 3, i+1)
    plt.imshow(img)
    plt.title(cat)
    plt.axis('off')
plt.show()
```

# **Le fine-tuning de DensNet169**

```python
# FINE-TUNING DenseNet169 pour détection de panneaux sales

import os
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import DenseNet169
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
import matplotlib.pyplot as plt

# 1 PARAMÈTRES GÉNÉRAUX
base_dir = "/kaggle/input/solar-panel-images/Faulty_solar_panel"  # chemin du dataset
img_size = (224, 224)
batch_size = 32
num_epochs = 15

# 2 DATA AUGMENTATION & CHARGEMENT DES IMAGES
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=25,
    width_shift_range=0.2,
    height_shift_range=0.2,
    zoom_range=0.2,
    shear_range=0.15,
    horizontal_flip=True,
    validation_split=0.2
)

train_gen = train_datagen.flow_from_directory(
    base_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='training'
)

val_gen = train_datagen.flow_from_directory(
    base_dir,
    target_size=img_size,
    batch_size=batch_size,
    class_mode='categorical',
    subset='validation'
)

num_classes = len(train_gen.class_indices)
print(f"Classes détectées : {train_gen.class_indices}")

# 3 CONSTRUCTION DU MODÈLE DenseNet169
base_model = DenseNet169(weights='imagenet', include_top=False, input_shape=(224, 224, 3))

# Débloquer toutes les couches pour fine-tuning complet
for layer in base_model.layers:
    layer.trainable = True

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dropout(0.4)(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.3)(x)
predictions = Dense(num_classes, activation='softmax')(x)

model = Model(inputs=base_model.input, outputs=predictions)

# 4 COMPILATION & CALLBACKS
model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

checkpoint = ModelCheckpoint("best_densenet_model.h5", monitor='val_accuracy', save_best_only=True, verbose=1)
early_stop = EarlyStopping(monitor='val_accuracy', patience=4, restore_best_weights=True)

# 5 ENTRAÎNEMENT
history = model.fit(
    train_gen,
    validation_data=val_gen,
    epochs=num_epochs,
    callbacks=[checkpoint, early_stop]
)

# 6 ÉVALUATION
val_loss, val_acc = model.evaluate(val_gen)
print(f"Validation Accuracy: {val_acc*100:.2f}%")

# 7 VISUALISATION DES COURBES
plt.figure(figsize=(8,5))
plt.plot(history.history['accuracy'], label='Train Acc')
plt.plot(history.history['val_accuracy'], label='Val Acc')
plt.title("Courbes d'apprentissage - DenseNet169 Fine-Tuning")
plt.xlabel("Epochs")
plt.ylabel("Accuracy")
plt.legend()
plt.show()

# 8 TEST SUR UNE IMAGE
test_class = list(train_gen.class_indices.keys())[0]
test_dir = os.path.join(base_dir, test_class)
test_img_path = os.path.join(test_dir, os.listdir(test_dir)[0])

img = tf.keras.preprocessing.image.load_img(test_img_path, target_size=img_size)
img_array = tf.keras.preprocessing.image.img_to_array(img)
img_array = np.expand_dims(img_array / 255.0, axis=0)

predictions = model.predict(img_array)
pred_class = list(train_gen.class_indices.keys())[np.argmax(predictions)]

plt.imshow(tf.keras.preprocessing.image.load_img(test_img_path))
plt.title(f"Prédiction : {pred_class} ({100 * np.max(predictions):.2f}%)")
plt.axis("off")
plt.show()
```
